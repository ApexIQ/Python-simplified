# Loops In Python

## For Loop

```python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

```python
# Calculating the total cost of items in a shopping cart

prices = [2.50, 1.00, 3.75, 0.79]
total_cost = 0

for price in prices:
    total_cost = total_cost + price

print(f"total cost: {total_cost}")
```

```python
# Checking if a bookshelf has enough space for new books

book_widths = [5, 8, 3, 6, 2, 4, 20]  # Widths of books in cm
shelf_width = 30  # Total shelf width in cm
current_width = 0

for width in book_widths:
  current_width = current_width + width 
  if current_width > shelf_width:
    print("Uh oh, not enough space!")
    break  # Stop checking if we run out of space
else:  # This runs only if the loop completes without a break
  print("You have space for more books!")
```

## Nested Loops - For

```python
for i in range(5):
    for j in range(i+1):
        print("*", end ="")
    print()
```

```python
# Creating a simple grid-based animation (using characters)

import time

for i in range(10):
  for j in range(20):
    if (i + j) % 2 == 0:
      print("#", end="")
    else:
      print("-", end="")
  print()
  time.sleep(0.5)  # Pause for half a second
```

```python
## While loop

count = 0
while count < 5:
    print(count)
    count += 1
```

```python
# Counting down until a rocket launch

countdown = 3
while countdown > 0:
  print(countdown, "...")
  countdown = countdown - 1

print("Blast off!")
```

```python
# Simulating a simple game where the user guesses a secret number

import random

secret_number = random.randint(1, 100)
guess = 0
num_guesses = 0

while guess != secret_number:
  guess = int(input("Guess a number between 1 and 100: "))
  num_guesses = num_guesses + 1

  if guess < secret_number:
    print("Too low!")
  elif guess > secret_number:
    print("Too high!")

print("You got it in", num_guesses, "guesses!")
```

```python
# nested loop - while + For

#Simulating a basic Conway's Game of Life (simplified)

import time
import random

# Initialize a grid with random 0s and 1s (0 = dead cell, 1 = live cell)
grid = [[random.randint(0, 1) for _ in range(20)] for _ in range(10)]

while True:  # Run indefinitely
  for row in grid:
    for cell in row:
      print(cell if cell else " ", end="")  # Print cell or space
    print()
  time.sleep(0.5)
```
